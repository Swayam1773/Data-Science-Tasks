```python
def maxmarks(groupname):
    k = df['race/ethnicity'] == groupname
    s = df[k]
    m = s['math score']
    r = s['reading score']
    w = s['writing score']
    mm = m.max()
    mr = r.max()
    mw = w.max()
    a = [mr,mw,mm]
    return [mr,mm,mw]
```
